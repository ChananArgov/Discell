```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

import seaborn as sns
import numpy as np
```

```python
"------------------------ Load Data -----------------------"

path = os.path.join('..', '..', 'Results', 'Multi_Cellular_Model_LOOCV_AUC.csv')
Multi_Cellular_Performance = pd.read_csv(path)
print(Multi_Cellular_Performance)
```

```python
"------------------------------- Plot ---------------------------------"
Multi_Cellular_Performance.rename(columns={'PR_AUC':'PR AUC', 'ROC_AUC': 'ROC AUC'}, inplace=True)
plt.figure(figsize=(15,10))

sns.scatterplot(data=Multi_Cellular_Performance, x="PR AUC", y="ROC AUC", hue="Cell_Type", s=100)

plt.yticks(np.arange(0.4, 1.1, 0.1))
plt.xticks(np.arange(0, 1.1, 0.1))
plt.axhline(y = 0.5, color = 'r', linestyle = '--')
plt.axvline(x = 0.1, color = 'black', linestyle = '--')

plt.legend(loc='best', fontsize='medium')#, ncols=3

path = os.path.join('..', '..', 'Results', 'Multi_Cellular_Model_Performance_Plot.pdf')
plt.savefig(path)
plt.show()
```
